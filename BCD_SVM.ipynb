```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report,roc_curve, roc_auc_score,auc
```

```python
data = pd.read_csv("/content/drive/MyDrive/breast-cancer.csv")
data
```

```python
data.isnull().sum()
```

```python
sensorWithInfo=[i for i in data.columns if 'Unnamed' in i]
data.drop(sensorWithInfo,axis=1,inplace=True)
data
```

```python
x=data.drop(['id','diagnosis'],axis=1)
y=data['diagnosis'].map({'M':1,'B':0})
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=SVC(probability=True)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred_prob=model.predict_proba(x_test)[:,1]
```

```python
print(f"Accuracy:\n{accuracy_score(y_test,y_pred)}")
print(f"Confusion Matrix:\n{confusion_matrix(y_test,y_pred)}")
print(f"Classification Report:\n{classification_report(y_test,y_pred)}")
```

```python
abc,bca,_=roc_curve(y_test,y_pred_prob)
roc_auc=auc(abc,bca)
print(f'area of curve:{roc_auc}')
```

```python
plt.figure(figsize=(8,6))
plt.plot(abc,bca,color='darkorange',lw=2,label='ROC Curve (area = %0.2f)' % roc_auc)
plt.plot([0,1],[0,1],color='navy',lw=2,linestyle='--')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

```python
import seaborn as sns
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,fmt='d',cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
